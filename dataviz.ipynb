```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
house = pd.read_csv('/content/Housing.csv')
house.head()
```

```python
# Checking for missing values
missing_values = house.isnull().sum()

print(missing_values)
```

```python
# Checking for duplicate rows
duplicate_rows = house.duplicated().sum()

print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
#histogram
plt.figure(figsize=(8, 6))
plt.hist(house['price'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Property Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()

```

```python
#bar plot
plt.figure(figsize=(8, 6))
sns.countplot(x='bedrooms', data=house, palette='Set2')
plt.title('Count of Properties by Number of bedrooms')
plt.xlabel('Number of bedrooms')
plt.ylabel('Count')
plt.show()
```

```python
#scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(house['area'], house['price'], color='purple')
plt.title('Relationship between area and price')
plt.xlabel('area')
plt.ylabel('price')
plt.show()
```

```python
#box plot
plt.figure(figsize=(8, 6))
sns.boxplot(x=house['price'], color='lightgreen')
plt.title('Boxplot of Property Prices')
plt.xlabel('Price')
plt.show()
```

```python
#pie chart
plt.figure(figsize=(8, 6))
house['furnishingstatus'].value_counts().plot.pie(autopct='%1.1f%%', colors=['lightblue', 'lightcoral', 'lightgreen'], explode=[0.05, 0.05, 0.05])
plt.title('Distribution of Furnishing Status')
plt.ylabel('')
plt.show()
```
